```python
def encode_date(df,column):
    df=df.copy()
    #converting to datetime
    df[column]=pd.to_datetime(df[column],format='%d-%m-%y')
    #extracting year,month and day
    df[column+'_year']=df[column].apply(lambda  x:x.year)
    df[column+'_month']=df[column].apply(lambda x:x.month)
    df[column+'_day']=df[column].apply(lambda x:x.day)
    df=df.drop(column,axis=1)
    return df
```

```python
def encode_duration(df,column):
    df=df.copy()
    df[column+'_yrs']=df[column].apply(lambda x:re.search(r'\d+(?=yrs)',x).group(0)).astype(int)
    df[column+'_mon']=df[column].apply(lambda x:re.search(r'\d+(?=mon)',x).group(0)).astype(int)
    df=df.drop(column,axis=1)
    
    return df
```

```python
def onehot_encode(df,column):
    df=df.copy()
    dummies=pd.get_dummies(df[column],prefix=column)
    df=pd.concat([df,dummies],axis=1)
    df=df.drop(column,axis=1)
    for column in df.columns:
        if df[column].dtypes=='bool':
            df[column]=df[column].astype(int)
    
    return df
```

```python
def preprocess_inputs(df):
    df=df.copy()
    #encoding missing employment value and unemploayed


    df['Employment.Type']=df['Employment.Type'].fillna('Unemployed')

    #dropping high cardinality column

    df=df.drop(['UniqueID','supplier_id','Current_pincode_ID','Employee_code_ID','MobileNo_Avl_Flag'],axis=1)

    df=encode_date(df,'Date.of.Birth')
    df=encode_date(df,'DisbursalDate')
    df=encode_duration(df,'AVERAGE.ACCT.AGE')
    df=encode_duration(df,'CREDIT.HISTORY.LENGTH')
    df=df.drop('DisbursalDate_year',axis=1)
    ONEHOT_COLUMNS=['branch_id','manufacturer_id','Employment.Type','State_ID','PERFORM_CNS.SCORE.DESCRIPTION']

    for column in ONEHOT_COLUMNS:
        df=onehot_encode(df,column)

    return df
```
